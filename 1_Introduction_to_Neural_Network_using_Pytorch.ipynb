Load a simple CSV file using Pandas

```python
import pandas as pd
df=pd.read_csv('diabetes.csv')
```

```python
df
```

Convert the dataframe to tensors

```python
import numpy as np
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_numpy=np.array(df.drop(['Outcome'],axis=1))
y_numpy=np.array(df['Outcome'])
y_numpy=y_numpy.reshape(-1,1)
n_samples, n_features = X_numpy.shape
X = torch.from_numpy(X_numpy.astype(np.float32))
y = torch.from_numpy(y_numpy.astype(np.float32))
```

Split the data set into training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

Implement a simple neural network with y=wx+b and subsequent sigmoid function at the output layer

```python
# 1) Model
# Linear model f = wx + b , sigmoid at the end
class Model(nn.Module):
    def __init__(self, n_input_features):
        super(Model, self).__init__()#already defined in Pytorch.nn, overriding this class
        self.l1 = nn.Linear(n_input_features, 20)
        self.l2=nn.Linear(20,1)


    def forward(self, x):
        x=torch.relu(self.l1(x))
        x = torch.sigmoid(self.l2(x))
        return x

model = Model(n_features)

```

Define the loss and optimizer used

```python
# 2) Loss and optimizer
num_epochs = 1000
learning_rate = 0.001
criterion = nn.BCELoss()#Binary cross entropy
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

Model Training

```python
# 3) Training loop
for epoch in range(num_epochs):
    # Forward pass and loss
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)

    # Backward pass and update
    loss.backward()
    optimizer.step()

    # zero grad before new step
    optimizer.zero_grad()

    if (epoch+1) % 10 == 0:
        print(f'epoch: {epoch+1}, loss = {loss.item():.4f}')
```

Model Testing

```python
model.eval()
y_pred = model(X_test)
test_loss = criterion(y_pred, y_test)
print('test loss is {}'.format(test_loss))
```
